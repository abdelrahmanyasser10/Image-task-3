```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def extractRed(img):
    hsv_lower = np.array([100, 100, 100])
    hsv_upper = np.array([140, 255, 255])
    
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    extracted_img = cv2.inRange(hsv_img, hsv_lower, hsv_upper)
    masked_img= cv2.bitwise_and(img, img, mask=extracted_img)
    return extracted_img, masked_img
```

```python
def gaussian_kernel(shape, sigma):
    r,c = shape
    x = np.linspace(-(r-1)/2, r-1, r)
    y = np.linspace(-(c-1)/2, c-1, c)

    X,Y = np.meshgrid(x,y)
    kernel = (1/(2*np.pi*np.square(sigma))) * np.exp(-0.5 * (np.square(X) + np.square(Y))/np.square(sigma))
    Kernel = kernel / np.sum(kernel)
    return Kernel  
```

```python
def GaussFilter(image, shape = (3,3), sigma = 3):
    gaussianKernel = gaussian_kernel(shape, sigma)
    blurred_image = cv2.filter2D(image, -1, gaussianKernel)
    return blurred_image
```

```python
img = cv2.imread('Task 1/Screenshot 2023-09-03 143627.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
extracted, RedLine = extractRed(img)

plt.imshow(extracted, cmap='gray')
plt.show()
plt.imshow(RedLine)
plt.show()
```

```python
def RedLineDetector(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY)
    circles = cv2.HoughCircles(thresh, cv2.HOUGH_GRADIENT, 1, 20,
                            param1=50, param2=30, minRadius=0, maxRadius=0)
    mask = np.ones(thresh.shape, dtype=np.uint8) * 255
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i in circles[0,:]:
            cv2.circle(mask, (i[0], i[1]), i[2], 0, -1)
    thresh = cv2.bitwise_and(thresh, thresh, mask=mask)
    thresh = cv2.bitwise_and(thresh, thresh, mask=mask)
    contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) > 0:
        c = max(contours, key=cv2.contourArea)
        x, y, w, h = cv2.boundingRect(c)
        distance_from_bottom = img.shape[0] - (y + h)
        img_with_rectangle = cv2.rectangle(img.copy(), (x, y), (x + w, y + h), (0, 255, 0), 3)
        img_with_rectangle = cv2.cvtColor(img_with_rectangle, cv2.COLOR_BGR2RGB)
        plt.imshow(img_with_rectangle)
        plt.show()
        return True, distance_from_bottom
    else:
        return False, None
```

```python
found_red_line, distance_from_bottom = RedLineDetector(RedLine)
if found_red_line:
    print(f'A red line was detected {distance_from_bottom} pixels from the bottom border of the image.')
else:
    print('No red line was detected in the image.')
```
